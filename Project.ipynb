```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from requests import get
print('Done!')
```

```python
url1 = 'http://www.usa.com/rank/new-york-state--population-density--zip-code-rank.htm'
response = get(url1)
print('Done!')
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
type(soup)
```

```python
column_names = ['Rank','Population Density','Zip Code']
df = pd.DataFrame(columns = column_names)
```

```python
table = soup.find('table', width="648")
```

```python
for tr_cell in table.find_all('tr'):
    row=[]
    for td_cell in tr_cell.find_all('td'):
        row.append(td_cell.text.strip())
        if len(row) == 3:
            df.loc[len(df)] = row
```

```python
df.head(100)
```

```python
df = df.rename(columns={'Population Density':'Population Density sq/mile'})
```

```python
df = df.drop([0])
```

```python
df.drop(['Rank'], axis = 1, inplace = True)
```

```python
df = df.reset_index()
```

```python
df = df.drop(['index'], axis =1)
```

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
address = 'New York, US'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_New_York = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, name, borough in zip(df_fin['Latitude'], df_fin['Longitude'], df_fin['Name'], df_fin['Borough']):
    label = '{}, {}'.format(name, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_New_York)  
    
map_New_York
```

```python
new_york_geo = 'nyc-zip-code-tabulation-areas-polygons.geojson' # geojson file

# create a plain world map
new_york_map = folium.Map(location=[latitude, longitude], zoom_start=10, tiles='Mapbox Bright')
```

```python
new_york_map.choropleth(
    geo_data = new_york_geo,
    data = df_fin,
    columns = ['Zip Code', 'Deaths'],
    key_on = 'feature.properties.postalCode',
    fill_color = 'YlOrRd', 
    fill_opacity = 0.7, 
    line_opacity = 0.2,
    legend_name = 'Number of Deaths'
)

# display map
new_york_map
```
